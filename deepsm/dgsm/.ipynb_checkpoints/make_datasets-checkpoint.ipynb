```python
import matplotlib as mpl 
import os
```

# Params

```python
resolution = 0.02
num_angle_cells = 56
min_radius = 0.3
max_radius = 5
radius_factor = 1.15
```

```python
angles = np.linspace(-180, 180, num_angle_cells+1)
print(angles)

r=min_radius
radiuses=[r]
v = 0.04
while r<max_radius:
    r = r+v
    radiuses.append(r)
    v*=radius_factor
    
radiuses = np.array(radiuses)
num_radius_cells = len(radiuses)-1
print(radiuses)
print(num_radius_cells)
```

# Display

```python
def plot_polar_scan(polar_scan):
    a, r = np.meshgrid(np.radians(angles), radiuses)
    ax = plt.subplot(111, projection='polar')
    ax.set_theta_zero_location("S")
    ax.set_theta_direction(-1)
    cmap = mpl.colors.LinearSegmentedColormap.from_list('mycmap', [(0, 'gray'),
                                                        (0.5, 'white'),
                                                        (1, 'black')])

    ax.pcolormesh(a, r, polar_scan.reshape(num_radius_cells, num_angle_cells) +1, cmap=cmap)
    plt.show()
```

### Load all sequences

```python
import pickle
def load_sequences(typ):
    seqdata = []
    for s in sorted(os.listdir(os.path.join(data_path, "polar_scans_"+typ))):
        fname = os.path.join(data_path, "polar_scans_"+typ, s)
        with open(fname, 'rb') as f:
            d = pickle.load(f)
            seqdata.extend(d)
    return seqdata
```

```python
real_data = load_sequences('real')
#testing_data = load_sequences('testing')
```

# Process room names

```python
import re
proc_real_data = []
for i in real_data:
    r_name = i[0]
    r_cat = i[1]
    if r_cat.lower() == 'none':
        r_name = re.sub("_.*", "", r_name)+"_door"
        r_cat = "door"
        proc_real_data.append([r_name, r_cat, i[2]])
    elif r_cat.lower() == "1p_office":
        r_cat = "small_office"
        proc_real_data.append([i[0], r_cat, i[2]])
    elif r_cat.lower() == "2p_office" or r_cat.lower() == "mp_office" or r_cat.lower() == "prof_office":
        r_cat = "large_office"
        proc_real_data.append([i[0], r_cat, i[2]])
    elif r_cat.lower() == "corridor":
        proc_real_data.append(i)
    elif (r_cat.lower() == "meeting_room" or r_cat.lower() == "elevator" 
          or r_cat.lower() == "kitchen" or r_cat.lower() == "living_room"
          or r_cat.lower() == "large_meeting_room"):
        proc_real_data.append(i)
```
